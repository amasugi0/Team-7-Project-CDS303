## Collaborating on code

Team collaboration on code is always a challenge. However, you are expected to do that here. I will teach you how to make it less of a challenge, because my class won't be the last time you'll have to do this.

- Have one team notebook, then let individuals have their own notebooks. One person must be responsible for the team notebook throughout the process.
- Use version control to keep track of the notebooks. If you don't you'll soon forget which notebook is the most recent. Do not rely on file dates.
- Each notebook should have specific tasks, inputs and outputs. An example is as follows:
    - book1: ingest, clean and transform data
    - book2: using the cleaned data from book1, perform exploratory data analysis and make further transformations as needed
    - book3: using the transformed data from book2, perform dimension reduction
    - book4: using the new components from book3, run some supervised learning algorithms and tune hyperparameters, test
    - teambook: run books 1-4, create outputs for a report or presentation
- This was an example, of course you can do this any way you like. The important things are to track versions and feature names, and have a main process the team follows. Also, ensure notebooks are updated at least weekly.

## How do combine notebooks easily and keep version references.

Let's say you have a folder where everyone puts their versions of book1, book2, book3, and book4. 

1. Specify a naming convention: book1_v1 so that the book name is first, followed by an underscore and the version.
2. Create a team notebook that integrates all the code and produces the final results, saved in the same folder.
3. Depending on how your code works, some things will need to be contained in one notebook because they won't transfer. Test everything to see if it works.


Start by importing the following:

```python
import glob
import shutil
```

Each time you want to update the team notebook with the latest releases, first grab a list of the current notebooks in the folder, by notebook name. For example, we'll get a list of all the current versions of each book, using 'glob'. Note, we could've done the following code recursively, rather than repetitively, but I did it this way to demonstrate more clearly what is going on. This example imports the wine dataset, cleans and preps it, splits it into train and test, runs several different regression models on it, and produces a table of all the runs and results.

```python
book_list1 = glob.glob('*book1*')  # The asterisk is a wildcard that will account for any characters before and after 'book1' 
book_list2 = glob.glob('*book2*')  # The asterisk is a wildcard that will account for any characters  before and after 'book2' 
book_list3 = glob.glob('*book3*')  # The asterisk is a wildcard that will account for any characters  before and after 'book3' 
book_list4 = glob.glob('*book4*')  # The asterisk is a wildcard that will account for any characters  before and after 'book4' 
```

Because the list will be in order, and your team has followed the naming conventions you specified, the latest release should be at the bottom of the list. To always pull the latest release, you'll use the following code to always pull the bottom item:

```python
i1 = len(book_list1) - 1     # we get the last item of a list by getting the length and subtracting 1, because all lists start indexing at 0
i2 = len(book_list2) - 1     # we get the last item of a list by getting the length and subtracting 1, because all lists start indexing at 0
i3 = len(book_list3) - 1     # we get the last item of a list by getting the length and subtracting 1, because all lists start indexing at 0
i4 = len(book_list4) - 1     # we get the last item of a list by getting the length and subtracting 1, because all lists start indexing at 0
```

Now that we've specified the file we want, we need to save it as the current version to be used in our code. To do that, we use 'shutil' to copy the file, in order to drop the version number. Note that shutil.copy will overwrite the file, while shutil.copyfile will throw an error if a file with the same name already exists:

```python
source = book_list1[i1]     # the file we're copying
destination = 'book_1_v1.ipynb'   # the new file without the version number, underscore differentiates from original file
print(source, destination)
_=shutil.copy(source, destination)  # by adding _= before shutil, we can suppress the annoying output 

source = book_list2[i2]     # the file we're copying
destination = 'book_2_v1.ipynb'   # the new file without the version number, underscore differentiates from original file
print(source, destination)
_=shutil.copy(source, destination)  # by adding _= before shutil, we can suppress the annoying output 

source = book_list3[i3]     # the file we're copying
destination = 'book_3_v1.ipynb'   # the new file without the version number, underscore differentiates from original file
print(source, destination)
_=shutil.copy(source, destination)  # by adding _= before shutil, we can suppress the annoying output 

ource = book_list4[i4]     # the file we're copying
destination = 'book_4_v1.ipynb'   # the new file without the version number, underscore differentiates from original file
print(source, destination)
_=shutil.copy(source, destination)  # by adding _= before shutil, we can suppress the annoying output 
```

Finally, we run the current version of each file, wherever it is appropriate to do so in our code:

```python
%run book_1_v1.ipynb            # now we can run the notebook file, and not worry about changing the code to account for the new version
```

```python
%run book_2_v1.ipynb            # now we can run the notebook file, and not worry about changing the code to account for the new version
```

```python
%run book_3_v1.ipynb            # now we can run the notebook file, and not worry about changing the code to account for the new version
```

```python
%run book_4_v1.ipynb            # now we can run the notebook file, and not worry about changing the code to account for the new version
```

Of course, you definitely want to save the team notebook as a new version every time you update it with new versions of your team's stuff.

Some things will automatically output in your team notebook, some will not. You can always call a dataframe or other thing in the team notebook to view it manually.
